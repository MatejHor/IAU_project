```python
def counter(column1, value, column2, color):
    count = 0
    for item in data[data[column1] == value][column2].str.contains(color):
        if item:
            count+=1 
    return count


print(counter("class", 1,"race","White"))
print(counter("class", 0,"race","White"))
print(counter("class", 1,"race","Black"))
print(counter("class", 0,"race","Black"))
```
