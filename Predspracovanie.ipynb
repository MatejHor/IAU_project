```python
import unittest

def secureMean(value1, value2, mean):
    if value1 is None or value1 is np.NaN or value1 == 0:
        if value2 is None or value2 is np.NaN or value2 == 0:
            return mean
        else:
            return value2
    elif value2 is None or value2 is np.NaN or value2 == 0:
        return value1
    else:
        return (value1 + value2)/2
    
class TestNotebook(unittest.TestCase):

    def test_secureMean(self):
         self.assertEqual(secureMean(-1,1,12),0)
         self.assertEqual(secureMean(3,3,12),3)
         # Test for 0
         self.assertEqual(secureMean(1,0,12),1)
         self.assertEqual(secureMean(0,2,12),2)
         self.assertEqual(secureMean(0,0,12),12)
         # Test for np.NaN
         self.assertEqual(secureMean(1,np.NaN,12),1)
         self.assertEqual(secureMean(np.NaN,2,12),2)
         self.assertEqual(secureMean(np.NaN,np.NaN,12),12)
         # Test for None
         self.assertEqual(secureMean(None,None,12),12)
         self.assertEqual(secureMean(0,None,12),12)
         self.assertEqual(secureMean(None,4,12),4)

unittest.main(argv=[''], verbosity=2, exit=False)
    
```

```python
def secureParser(value1, value2, mode):
    if value1 == None or value1 == '' or value1 == '?':
        if value2 == None or value2 == '' or value2 == '?':
            return mode
        else:
            return value2
    elif value2 == None or value2 == '' or value2 == '?':
        return value1
    else:
        return value1

class TestNotebook(unittest.TestCase):

    def test_secureParser(self):
         self.assertEqual(secureParser('first','second','value'),'first')
         # Test for '' 
         self.assertEqual(secureParser('','','value'),'value')
         self.assertEqual(secureParser('first','','value'),'first')
         self.assertEqual(secureParser('','second','value'),'second')         
         # Test for '?'
         self.assertEqual(secureParser('first','?','value'),'first')
         self.assertEqual(secureParser('?','second','value'),'second')
         self.assertEqual(secureParser('?','?','value'),'value')
         # Test for None
         self.assertEqual(secureParser(None,'second','value'),'second')
         self.assertEqual(secureParser('first',None,'value'),'first')
         self.assertEqual(secureParser(None,None,'value'),'value')

unittest.main(argv=[''], verbosity=2, exit=False)
```
